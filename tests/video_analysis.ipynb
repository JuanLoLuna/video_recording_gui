```python
import cv2
import matplotlib.pyplot as plt

video_path = "/Users/jl4459/Library/CloudStorage/Box-Box/SmartSleeve/Repositories/video_recording/tests/raw_videos/recording_20251111_175410.avi-0000.avi"

cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise RuntimeError(f"Could not open video: {video_path}")

# Video-level metadata
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps         = cap.get(cv2.CAP_PROP_FPS)
width       = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height      = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print("Video info:")
print("  Frame count:", frame_count)
print("  FPS        :", fps)
print("  Size       :", width, "x", height)
```

```python
import csv
import numpy as np

csv_path = "/Users/jl4459/Library/CloudStorage/Box-Box/SmartSleeve/Repositories/video_recording/tests/raw_videos/recording_20251111_175410_metadata.csv"

rows = []
with open(csv_path, newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        rows.append(row)

print("Total rows:", len(rows))
print("First row:", rows[0])

# Convert to NumPy arrays
record_idx = np.array([int(r["record_frame_index"]) for r in rows])
frame_id   = np.array([int(r["camera_frame_id"]) for r in rows])
ts_us      = np.array([int(r["timestamp_us"]) for r in rows])
system_t   = np.array([float(r["system_time"]) for r in rows])

# Example: check timing
dt = np.diff(ts_us) / 1e9  # seconds between frames
print("Mean dt (s):", dt.mean())
print("Approx FPS :", 1.0 / dt.mean())
```
